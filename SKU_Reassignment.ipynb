```python
# Import dependancies and reshape the original df
import pandas as pd
import numpy as np
df = pd.read_excel('SKU_Master3-26.xlsx', header=0)
df = df.iloc[:,0:10]
df.head(30)
```

```python
# Isolate Individual artist and sort by ascending inventory ID
Kim_Frances_df = df[df['Artist']=='Kim Frances']
Kim_Frances_df = Kim_Frances_df.sort_values(by='INVENTORY ID', ascending=True)
Kim_Frances_df.head()
```

For each unique Description, there should be a different SKU #, regardless of the frame. This will account for the varying frame types.
Also, though, for like Descriptions, the substrate might vary.

```python
# Assuming your DataFrame is named df
# Filter DataFrame by unique values in 'Description' column and subatrate col
filtered_df = Kim_Frances_df.groupby('Description').apply(lambda x: x.drop_duplicates('Substrate'))

# Reset index to flatten the DataFrame
filtered_df = filtered_df.reset_index(drop=True)

# Display the filtered DataFrame
filtered_df
```

```python
# Create a list of values from 'GP2001701' to 'GP2001731'
new_skus = [f"GP20017{i:02d}" if int(i) < 10 else f"GP20017{i}" for i in range(1, 32)]

# Create a DataFrame with a column named 'New_Column' and assign the new values

filtered_df['New_SKUs'] = new_skus

# Display the DataFrame
filtered_df
```

```python
# Extract the base part of the inventory id (before the dash)
filtered_df['Old Skus'] = filtered_df['INVENTORY ID'].str.split('-', expand=True)[0]

# Display the DataFrame with the new column
filtered_df
```

```python
# Iterate through each value in 'Old Skus' column
dropped_skus=[]

for index, new_sku in enumerate(filtered_df['New_SKUs']):
    # Check if the value is in 'New_SKUs' column
    if new_sku in filtered_df['Old Skus'].values:
        # Remove the value from both 'Old Skus' and 'New_SKUs' columns
        dropped_skus.append(new_sku)
        filtered_df.at[index, 'New_SKUs'] = np.nan

# Display the updated DataFrame
filtered_df
```
